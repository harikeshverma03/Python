```python
%matplotlib inline
from pylab import *
import numpy as np

np.random.seed(2)
pageSpeed = np.random.normal(3.0, 1.0, 1000)
purchase = np.random.normal(50.0, 10.0, 1000) / pageSpeed
scatter(pageSpeed, purchase)
plt.show
```

```python
x = np.array(pageSpeed)
y = np.array(purchase)

p2 = np.poly1d(np.polyfit(x,y,7))
```

```python
import matplotlib.pyplot as plt

xp = np.linspace(0, 8 ,100)
plt.scatter(x, y)
plt.plot(xp, p2(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, p2(x))
print(r2)
```
