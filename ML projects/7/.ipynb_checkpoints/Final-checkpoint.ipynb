# Step - 1
Using the python methods deserialize the json file into a usable python object named RedliningData

```python
from urllib.request import urlopen
import json
url = "https://dsl.richmond.edu/panorama/redlining/static/downloads/geojson/MIDetroit1939.geojson"
response = urlopen(url)
RedliningData = json.loads(response.read())
```

# Step - 2
Define a class DetroitDisrtrict then use list comprehension to make 238 objects of DetroitDistrict class in a variable named Districts

```python
class DetroitDistrict():
    RandomLat = None
    RandomLong = None
    MedianIncome = None
    CensusTract = None
    def __init__(self, HolcGrade, HolcColor, name, Qualitative_des, Coordinates = []):
        self.Coordinates = Coordinates
        self.HolcGrade = HolcGrade
        self.HolcColor = HolcColor
        self.name = name
        self.Qualitative_des = Qualitative_des
```

```python
Districts = []
for i in RedliningData['features']:
    Cord = i['geometry']['coordinates'][0]
    HolcG = i['properties']['holc_grade']
    if(HolcG == 'A'):
        HolcColor = 'darkgreen'
    elif(HolcG == 'B'):
        HolcColor = 'cornflowerblue'
    elif(HolcG == 'C'):
        HolcColor = 'gold'
    elif(HolcG == 'D'):
        HolcColor = 'maroon'
    name = i['properties']['area_description_data']['9']
    area_des = i['properties']['area_description_data']['8']
    temp = DetroitDistrict(HolcG, HolcColor, name, area_des, Cord)
    Districts.append(temp)
```

# Step - 3
Using matplotlib documentation complete the cells to get desired plot

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
fig, ax = plt.subplots()
for i in Districts: # what kind of for loop makes sense?
    ax.add_patch(plt.Polygon(i.Coordinates[0], facecolor = i.HolcColor, edgecolor = 'black')) # add arguments here
    ax.autoscale()
    plt.rcParams["figure.figsize"] = (15,15)
plt.show()
```

# Step - 4
Getting a clear understanding of what the code does and comment important steps

```python
import random as random
from matplotlib.path import Path
import numpy as np
random.seed(17) #setting random seed to integer to get reproducible results
xgrid = np.arange(-83.5,-82.8,.004) #creating an array with values from -83.5 to -82.8 with 0.004 division
ygrid = np.arange(42.1, 42.6, .004)
xmesh, ymesh = np.meshgrid(xgrid,ygrid)#creating a mest using x and y coordinates
points = np.vstack((xmesh.flatten(),ymesh.flatten())).T #flattening the mesh to obtain all the points in the mesh
for j in Districts:
    p = Path(j.Coordinates[0]) #creates an outline of the coordinates
    grid = p.contains_points(points) #getting the values from the mesh which lies in the district 
    print(j," : ", points[random.choice(np.where(grid)[0])]) #rondomly choosing a value lying in the grid
    point = points[random.choice(np.where(grid)[0])]
    j.RandomLong = point[0] #setting random longitude fromm chosen point
    j.RandomLat = point[1] #setting random lattitude from chosen point
```

# Step-5
use the api https://geo.fcc.gov/api/census/ to get the census tract information of each district

```python
import requests
for i in Districts:
    parameters = {
        "lat":i.RandomLat,
        "lon":i.RandomLong,
        "censusYear": 2020
    }
    response = requests.get('https://geo.fcc.gov/api/census/area?', params=parameters)
    a = response.json()
    print(a['results'][0]['block_fips'])
    i.CensusTract = a['results'][0]['block_fips']
```

# Step-6
Using the https://api.census.gov/data/2018/acs/acs5? api to get the MEDIAN HOUSEHOLD INCOME IN THE PAST 12 MONTHS (IN 2015 INFLATION-ADJUSTED
DOLLARS)

```python
for i in Districts:
    st = 'state:'+ i.CensusTract[0:2]
    cou = ' county:'+i.CensusTract[2:5]
    tra = ' tract:'+i.CensusTract[5:11]
    block = 'block group:'+i.CensusTract[11:12]
    parameters = {
        "for":block,
        "in":st+cou+tra,
        "key":"292a25619c2b08fc6c50407e227de4f6664e962b",#generated using emaild id
        "get":"B19013_001E" #key to get median household income in past 12 months for 2018
    }
    while True:
        try:
            response = requests.get('https://api.census.gov/data/2018/acs/acs5?', params = parameters)
        except Exception:#To cope up with the error when the site is sometimes unresponsive
            continue;
        else:
            break
    try:
        c = response.json()
    except ValueError:#To check if the data is available for tract or not
        print("No Response")
        i.MedianIncome = 0
        continue;
    if(c[1][0] == None):#If the Median Income is None 0 is chosen as Median Income for that district, it will not be counted while taking average
        i.MedianIncome = 0
    else:
        i.MedianIncome = int(c[1][0])
    print(i.MedianIncome)
```

# Step-7
For each district grade (A, B, C, D). Calculate the mean and median of the median household income
data that you gathered.

```python
import statistics
Inc_A = []
Inc_B = []
Inc_C = []
Inc_D = []
for i in Districts:
    #i.MedianIncome > 0 it is used as some responses whose values are not available are set to -666666666
    if(i.HolcGrade == 'A' and i.MedianIncome > 0):
        Inc_A.append(i.MedianIncome)
    elif(i.HolcGrade == 'B'and i.MedianIncome > 0):
        Inc_B.append(i.MedianIncome)
    elif(i.HolcGrade == 'C'and i.MedianIncome > 0):
        Inc_C.append(i.MedianIncome)
    elif(i.HolcGrade == 'D'and i.MedianIncome > 0):
        Inc_D.append(i.MedianIncome)
A_mean_income = statistics.mean(Inc_A)
B_mean_income = statistics.mean(Inc_B)
C_mean_income = statistics.mean(Inc_C)
D_mean_income = statistics.mean(Inc_D)
A_median_income = statistics.median(Inc_A)
B_median_income = statistics.median(Inc_B)
C_median_income = statistics.median(Inc_C)
D_median_income = statistics.median(Inc_D)
print(A_mean_income, A_median_income)
print(B_mean_income, B_median_income)
print(C_mean_income, C_median_income)
print(D_mean_income, D_median_income)
```

# Step-7
Use a list comprehension or other method to combine all the qualitative description strings for each
district category. Find the 10 most common words that are unique to each category (meaning don’t include words that
are common to all 4 categories).

```python
#Step 7 to find 10 most common words
import pandas as pd
import re
Qua_A = ''
Qua_B = ''
Qua_C = ''
Qua_D = ''
for i in Districts:
    if(i.HolcGrade == 'A'):
        Qua_A = Qua_A + (i.Qualitative_des)
    elif(i.HolcGrade == 'B'):
        Qua_B = Qua_B + (i.Qualitative_des)
    elif(i.HolcGrade == 'C'):
        Qua_C = Qua_C + (i.Qualitative_des)
    elif(i.HolcGrade == 'D'):
        Qua_D = Qua_D + (i.Qualitative_des)
res_A = re.split(r"[\b\W\b]+", Qua_A) #using regex operation to split the string into words
res_B = re.split(r"[\b\W\b]+", Qua_B) #using regex operation to split the string into words
res_C = re.split(r"[\b\W\b]+", Qua_C) #using regex operation to split the string into words
res_D = re.split(r"[\b\W\b]+", Qua_D) #using regex operation to split the string into words     
unique_words = set(res_A)
unique_words_A = set(res_A)
unique_words_B = set(res_B)
unique_words_C = set(res_C)
unique_words_D = set(res_D)
#words common in all the categories are removed
for words in unique_words:
    if(words in unique_words_B and words in unique_words_C and words in unique_words_D):
        unique_words_A.remove(words)
        unique_words_B.remove(words)
        unique_words_C.remove(words)
        unique_words_D.remove(words)
        print(words)  
```

```python
#10 most common words for A
liA = []
for words in unique_words_A:
    liA.append([words, res_A.count(words)])
df = pd.DataFrame(liA, columns = ['A', 'num_A'])
df = df.sort_values(by=['num_A'], ascending = False)
A_10_Most_Common = []
for i, r in df.iterrows():
    if(len(A_10_Most_Common) < 10):
        A_10_Most_Common.append(r[0])
A_10_Most_Common
```

```python
#10 most common words for B
liB = []
for words in unique_words_B:
    liB.append([words, res_B.count(words)])
df = pd.DataFrame(liB, columns = ['B', 'num_B'])
df = df.sort_values(by=['num_B'], ascending = False)
B_10_Most_Common = []
for i, r in df.iterrows():
    if(len(B_10_Most_Common) < 10):
        B_10_Most_Common.append(r[0])
B_10_Most_Common
```

```python
#10 most common words for C
liC = []
for words in unique_words_C:
    liC.append([words, res_C.count(words)])
df = pd.DataFrame(liC, columns = ['C', 'num_C'])
df = df.sort_values(by=['num_C'], ascending = False)
C_10_Most_Common = []
for i, r in df.iterrows():
    if(len(C_10_Most_Common) < 10):
        C_10_Most_Common.append(r[0])
C_10_Most_Common
```

```python
#10 most common words for D
liD = []
for words in unique_words_D:
    liD.append([words, res_D.count(words)])
df = pd.DataFrame(liD, columns = ['D', 'num_D'])
df = df.sort_values(by=['num_D'], ascending = False)
D_10_Most_Common = []
for i, r in df.iterrows():
    if(len(D_10_Most_Common) < 10):
        D_10_Most_Common.append(r[0])
D_10_Most_Common
```
