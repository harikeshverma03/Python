```python
class DetroitDistrict():
    RandomLat = None
    RandomLong = None
    MedianIncome = None
    CensusTract = None
    def __init__(self, HolcGrade, HolcColor, name, Qualitative_des, Coordinates = []):
        self.Coordinates = Coordinates
        self.HolcGrade = HolcGrade
        self.HolcColor = HolcColor
        self.name = name
        self.Qualitative_des = Qualitative_des
```

```python
Districts = []
for i in RedliningData['features']:
    Cord = i['geometry']['coordinates'][0]
    HolcG = i['properties']['holc_grade']
    if(HolcG == 'A'):
        HolcColor = 'darkgreen'
    elif(HolcG == 'B'):
        HolcColor = 'cornflowerblue'
    elif(HolcG == 'C'):
        HolcColor = 'gold'
    elif(HolcG == 'D'):
        HolcColor = 'maroon'
    name = i['properties']['area_description_data']['9']
    area_des = i['properties']['area_description_data']['8']
    temp = DetroitDistrict(HolcG, HolcColor, name, area_des, Cord)
    Districts.append(temp)
```
